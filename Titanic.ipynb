```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r"D:\\ML data\titanic\train.csv")
```

```python
df.head()
```

```python
df=df.drop(['Name','Ticket','Fare','Cabin','Embarked','PassengerId'],axis =1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
df['Sex'] = LabelEncoder().fit_transform(df['Sex'])
df.head()
```

```python
mean = df['Age'].mean()
df['Age'].fillna(value = mean , inplace = True)
```

```python
xtrain = df.iloc[:, 1:]
ytrain = df.iloc[:, 0]
xtrain , ytrain
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(xtrain , ytrain)
```

```python
ypred = classifier.predict(xtrain)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytrain , ypred)
cm
```

```python
ac = (cm[0,0] + cm[1,1])/(cm[0,0] + cm[1,1]+cm[1,0]+cm[0,1])
ac
```
